```python
### 데이터 업로드 함수
def upload_data(file_name):
  rtn = pd.read_csv(file_name, encoding = 'cp949')
  rtn = pd.DataFrame(rtn)

  return rtn
```

```python
park_data = upload_data("/content/drive/MyDrive/Colab Notebooks/헤도닉_집값예측/x_data/편의시설/경기도_안양시_도시공원정보_20221115.csv")
lib_data = upload_data("/content/drive/MyDrive/Colab Notebooks/헤도닉_집값예측/x_data/편의시설/경기도공공도서관현황.csv")
```

```python
## 공원 데이터 불필요한 부분 쳐내기
park_data.dropna(subset=['공원구분'], inplace = True)
park_data = park_data[(park_data['공원구분'] != '어린이공원') & (park_data['공원구분'] != '소공원')]

park_df = park_data[['공원명', '공원구분', '소재지지번주소', '위도', '경도']]

# 공원데이터 최종
park_df
```

```python
# 도서관 데이터 불필요한 부분 쳐내기
lib_data = lib_data[lib_data['시군명'].str.contains('안양')]
lib_data.rename(columns = {'WGS84위도': '위도', 'WGS84경도': '경도'}, inplace=True)

lib_df = lib_data[['도서관명','소재지주소','위도','경도']]
```

```python
park_df.to_csv("공원데이터.csv", index=False, encoding='utf-8-sig')
lib_df.to_csv("도서관데이터.csv", index=False, encoding='utf-8-sig')
```
