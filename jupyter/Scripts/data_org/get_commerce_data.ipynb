```python
# 패키지 import 구간
import json
import requests
import pandas as pd
import xml.etree.ElementTree as ET
import requests
import datetime
import math

from bs4 import BeautifulSoup as BS
from google.colab import drive

from geopy.geocoders import Nominatim
```

```python
drive.mount('/content/drive')
google_maps_API = ""
```

```python
# 데이터 업로드
def upload_data(file_name):
  rtn = pd.read_csv(file_name, encoding = 'utf-8')
  rtn = pd.DataFrame(rtn)

  return rtn


# Google Map api 위도, 경도 추출 함수
def get_lat_lng(address, api_key):
    endpoint = f'https://maps.googleapis.com/maps/api/geocode/json?address={address}&key={api_key}'
    response = requests.get(endpoint)
    data = response.json()
    if data['status'] == 'OK':
        lat = data['results'][0]['geometry']['location']['lat']
        lng = data['results'][0]['geometry']['location']['lng']
        return lat, lng
    else:
        return 0, 0


# 지번 주소를 기반으로 위도 경도 추출
def getLocList(data):
  latitude = []
  longitude = []

  for i in data['번지주소']:
    lat, lng = get_lat_lng(i, google_maps_API)
    latitude.append(lat)
    longitude.append(lng)

  loc = pd.DataFrame({'위도' : latitude, '경도' : longitude})
  rtn = pd.merge(data, loc, left_index = True, right_index = True, how = 'inner')

  return rtn
```

```python
mart_data = upload_data("/content/drive/MyDrive/Colab Notebooks/헤도닉_집값예측/x_data/상업시설데이터.csv")
mart_data = mart_data[['사업장명', '소재지전체주소']]

### mart_data가 nan값인 쪽이 있어서 일단 누락 시킴
mart_data.rename(columns = {'소재지전체주소': '번지주소'}, inplace = True)

mart_df = getLocList(mart_data)
```
