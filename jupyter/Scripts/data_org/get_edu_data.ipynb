```python
### 데이터 업로드 함수
def upload_data(file_name):
  rtn = pd.read_csv(file_name, encoding = 'cp949')
  rtn = pd.DataFrame(rtn)

  return rtn
```

```python
# 고등, 중등, 초등, 유치원 --> 기본 교육 시설
high_data = upload_data("/content/drive/MyDrive/Colab Notebooks/헤도닉_집값예측/x_data/교육시설/전처리이전데이터/고등학교현황.csv")
middle_data = upload_data("/content/drive/MyDrive/Colab Notebooks/헤도닉_집값예측/x_data/교육시설/전처리이전데이터/중학교현황.csv")
elem_data = upload_data("/content/drive/MyDrive/Colab Notebooks/헤도닉_집값예측/x_data/교육시설/전처리이전데이터/초등학교현황.csv")
kinder_data = upload_data("/content/drive/MyDrive/Colab Notebooks/헤도닉_집값예측/x_data/교육시설/전처리이전데이터/유치원현황.csv")

# 학원 --> 선택 교육 시설
academy_data = upload_data("/content/drive/MyDrive/Colab Notebooks/헤도닉_집값예측/x_data/교육시설/전처리이전데이터/학원현황.csv")
```

```python
high_data_coordinate = high_data[['시설명','소재지도로명주소','WGS84위도','WGS84경도']]
middle_data_coordinate = middle_data[['시설명','소재지도로명주소','WGS84위도','WGS84경도']]
elem_data_coordinate = elem_data[['시설명','소재지도로명주소','WGS84위도','WGS84경도']]
kinder_data_coordinate = kinder_data[['유치원명','정제도로명주소','정제WGS84위도','정제WGS84경도']]
academy_data_coordinate = academy_data[['시설명','소재지도로명주소','WGS84위도','WGS84경도']]
```

```python
kinder_data_coordinate.dropna(subset=['정제도로명주소'], inplace = True)
kinder_data_coordinate = kinder_data_coordinate[kinder_data_coordinate['정제도로명주소'].str.contains('수원시')]

high_data_coordinate.dropna(subset=['소재지도로명주소'], inplace = True)
high_data_coordinate = high_data_coordinate[high_data_coordinate['소재지도로명주소'].str.contains('수원시')]

middle_data_coordinate.dropna(subset=['소재지도로명주소'], inplace = True)
middle_data_coordinate = middle_data_coordinate[middle_data_coordinate['소재지도로명주소'].str.contains('수원시')]

elem_data_coordinate.dropna(subset=['소재지도로명주소'], inplace = True)
elem_data_coordinate = elem_data_coordinate[elem_data_coordinate['소재지도로명주소'].str.contains('수원시')]

academy_data_coordinate.dropna(subset=['소재지도로명주소'], inplace = True)
academy_data_coordinate = academy_data_coordinate[academy_data_coordinate['소재지도로명주소'].str.contains('수원시')]
```

```python
kinder_data_coordinate.columns = ['유치원명','도로명주소','위도','경도']
high_data_coordinate.columns = ['고등학교명','도로명주소','위도','경도']
middle_data_coordinate.columns = ['중학교명','도로명주소','위도','경도']
elem_data_coordinate.columns = ['초등학교명','도로명주소','위도','경도']
academy_data_coordinate.columns = ['학원명','도로명주소','위도','경도']
```
