```python
### 데이터 업로드 함수
def upload_data(file_name):
  rtn = pd.read_csv(file_name, encoding = 'cp949')
  rtn = pd.DataFrame(rtn)

  return rtn
```

```python
# 의료시설 데이터 로드
hospital_data = upload_data("/content/drive/MyDrive/Colab Notebooks/헤도닉_집값예측/전처리이전데이터/병원현황(병원급).csv")
clinic_data = upload_data("/content/drive/MyDrive/Colab Notebooks/헤도닉_집값예측/전처리이전데이터/의원현황(의원급).csv")
general_data_origin = upload_data("/content/drive/MyDrive/Colab Notebooks/헤도닉_집값예측/전처리이전데이터/종합병원현황(병원급).csv")
```

```python
# 종합병원 데이터 결측값 제거
general_data_origin.dropna(subset=['시군명'], inplace = True)
general_data_origin.dropna(subset=['영업상태명'], inplace = True)

# 안양시 데이터만 추출
general_data_origin = general_data_origin[general_data_origin['시군명'].str.contains('안양')]

# 대형 병원 데이터만 추출
general_data = general_data_origin[general_data_origin['영업상태명'].str.contains('운영중')]

# 종합 병원 데이터에서 약국 데이터만 추출
general_data_open = general_data_origin[general_data_origin['영업상태명'].str.contains('영업중')]
pharmacy_data = general_data_open[general_data_open['사업장명'].str.contains('약국')]
```

```python
# 의원 데이터 결측값 제거
clinic_data.dropna(subset=['시군명'], inplace = True)
clinic_data.dropna(subset=['영업상태명'], inplace = True)

# 안양시 의원 데이터만 추출
clinic_data = clinic_data[clinic_data['시군명'].str.contains('안양')]
clinic_data = clinic_data[clinic_data['영업상태명'].str.contains('운영중')]
```

```python
# 일반 병원 데이터 결측값 제거
hospital_data.dropna(subset=['시군명'], inplace = True)
hospital_data.dropna(subset=['영업상태명'], inplace = True)

# 안양시 일반 병원 데이터만 추출
hospital_data = hospital_data[hospital_data['시군명'].str.contains('안양')]
hospital_data = hospital_data[hospital_data['영업상태명'].str.contains('운영중')]
```

```python
hospital_data_coordinate = hospital_data[['사업장명','소재지도로명주소','WGS84위도','WGS84경도']]
clinic_data_coordinate = clinic_data[['사업장명','소재지도로명주소','WGS84위도','WGS84경도']]
pharmacy_data_coordinate = pharmacy_data[['사업장명','소재지도로명주소','WGS84위도','WGS84경도']]
general_data_coordinate = general_data[['사업장명','소재지도로명주소','WGS84위도','WGS84경도']]
```

```python
hospital_data_coordinate
```

```python
hospital_data_coordinate.rename(columns = {'WGS84위도' : '위도', 'WGS84경도' : '경도'}, inplace=True)
clinic_data_coordinate.rename(columns = {'WGS84위도' : '위도', 'WGS84경도' : '경도'}, inplace=True)
pharmacy_data_coordinate.rename(columns = {'WGS84위도' : '위도', 'WGS84경도' : '경도'}, inplace=True)
general_data_coordinate.rename(columns = {'WGS84위도' : '위도', 'WGS84경도' : '경도'}, inplace=True)

hospital_data_coordinate.to_csv('병원데이터.csv', index=False, encoding='utf-8-sig')
clinic_data_coordinate.to_csv('의원데이터.csv', index=False, encoding='utf-8-sig')
pharmacy_data_coordinate.to_csv('약국데이터.csv', index=False, encoding='utf-8-sig')
general_data_coordinate.to_csv('종합병원데이터.csv', index=False, encoding='utf-8-sig')
```
